#### 컨슈머는 spark쪽
#### 카프카클러스터로 데이터를 받아서 

```python
import sys
import os
import sqlite3
from pyspark import SparkContext, SparkConf
from pyspark.streaming import StreamingContext # 실시간 데이터를 처리하기 위해서 StreamingContext
from pyspark.streaming.kafka import KafkaUtils
```

```python
# 카프카 클러스터는 디폴트로 라이브러리 패키지가 없기 때문에
# 다운을 직접 받음
os.environ['PYSPARK_SUBMIT_ARGS'] = '--packages org.apache.spark:spark-streaming-kafka-0-8_2.11:2.3.2 pyspark-shell'
```

```python
sc = SparkContext(appName='streamTest')
ssc = StreamingContext(sc, 5) # 스트리밍 컨텍스트 객체 생성 # 카프카 클러스터
# 5초 단위로 해당 토픽이 들어왔는지 폴링을 해서 데이터가 들어왔으면 컨슈머가 가져감
```

```python
# DB에 데이터 넣는 함수

def insertData(data):
    sql = "insert into ktest values(?,?)" # 튜플로 들어감
    db = sqlite3.connect(r'C:\Users\master16\Documents\my.db')
    db.execute( sql, data ) # ktest테이블에 data를 넣어줌
    db.commit()# insert, delete, update
    db.close()
```

```python
# 잘 불러왔는지 확인하는 함수

# 토픽이 있으면 콜백함수가 호출됨
def recvData(data):
    insertData(data) # 함수 호출
    print("data=", data) # ('키', '값') 튜플
    return data
```

```python
topic = 'sample'
# 내가 관심있는 토픽은 sample이다!
kvs = KafkaUtils.createStream(ssc, 'localhost', 'raw-event-streaming-consumer', {topic:1} )
# 포트는 생략하면 디폴트 포트
# sample토픽을 받겠다, 토픽이 하나의 파티션을 사용한다!

#=======================================
#메시지 수신시 recvData call
lines = kvs.map( recvData ) # 콜백함수
lines.pprint() #상태메시지...
#=======================================
ssc.start() # 폴링 시작(5초에 한 번씩 카프카클러스터에 sample topic확인)
ssc.awaitTerminationOrTimeout(2) # 2초 정도 대기했다가 중지하겠다고 등록! 
```

```python
ssc.stop() # 폴링중지
```
