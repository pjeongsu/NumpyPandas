# cost(MSE) test

```python
def cost(x, y, w):
    c = 0
    for i in np.arange(len(x)):
        hx = w*x[i] # 예측값
        c = c + (hx - y[i])**2 # 잔차 제곱의 누적합
    return c/len(x)
        
```

## 방법2 : 학습

```python
# cost함수 => 2차 함수

def cost(x, y, w):
    c = 0
    for i in np.arange(len(x)):
        hx = w*x[i] # 예측값
        c = c + (hx - y[i])**2 # 잔차 제곱의 누적합
    return c/len(x)
```

```python
# def cost(x, y, w):
#     c = 0
#     for i in np.arange(len(x)):
#         hx = w*x[i] # 예측값
#         c = c + (hx - y[i])**2 # 잔차 제곱의 누적합
#     return c/len(x)


# GD(Gradient descent, 경사하강법)
# cost에 대한 미분값 -> 위의 그래프에서 기울기를 구해줌

# 접선의 기울기를 구하는 함수 gradient_descent -> w 값이 있을 때 해당 값에서의 기울기

def gradient_descent( x,y,w ):
    c = 0
    for i in range(len(x)):
        hx = w*x[i] # 예측값
        c = c + (hx-y[i])*x[i] #편미분 # 합성함수의 미분
    return c/len(x)
```

```python
def show_gradient(x, y, w):
    for i in range(200): # 학습횟수
        c = cost(x, y, w)
        g = gradient_descent(x, y, w)
        print(i,c,w,g) # 학습횟수, cost(MSE), w, cost함수에서의 기울기
        w = w-0.1*g # 0.1(learning rate)*기울기 # w 초기값을 주고 기울기를 줄여나가면서 
    print("w :", w)
```

```python
show_gradient(x_data, y_data, 10)
```

## 방법1 : 공식

```python
import scipy.stats as st
```

```python
result = st.linregress(df['english'], df['mathematics'])
result
```

```python
print('기울기 :', result.slope) # w
print('절편 :', result.intercept)
print('상관계수 :', result.rvalue)
print('p-value :', result.pvalue)
print('표준편차 :', result.stderr)
```

```python
# 귀무가설 : 영어점수와 수학점수 간에 연관성이 없다. (기울기 : 0)
# 대립가설 : 기울기가 0이 아니다(영어점수와 수학점수)
# 회귀계수값(w, b) 의미가 있냐 없냐 
# -> 0.05보다 작으면 귀무가설 기각
# -> 0.05보다 크면 귀무가설 채택
# 결론 : 회귀계수가 예측값으로 활용할 수 있다.
```

```python
# 영어점수가 70점인 경우 수학점수를 예측하시오.
result.slope*70 + result.intercept
```

```python
plt.xlabel('english')
plt.ylabel('math')
plt.scatter(df['english'], df['mathematics'])
lm = result.slope*df['english'] + result.intercept
plt.plot(df['english'],lm, 'r--')
plt.show()
```
