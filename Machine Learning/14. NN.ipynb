- 서로 같으면 False
- 하나라도 다르면 True

```python
import numpy as np 
import pandas as pd 
import seaborn as sb 
import matplotlib.pyplot as plt
import mglearn
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.linear_model import LogisticRegression
import warnings
warnings.simplefilter('ignore')
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]]) # 특성데이터 2개 (4X2)
y_data = np.array( [[0],[1],[1],[0]]) # 이진분류
```

```python
model_logi = LogisticRegression(max_iter=1000, verbose=1) 
model_logi.fit(x_data, y_data)
```

```python
model_logi.coef_ # 가중치 2개
```

```python
model_logi.score( x_data, y_data )
```

```python
model_logi.predict( x_data ) # xor 문제 => 에측을 못함
```

### 중간 레이어 계층은 relu를 통과하고 마지막에는 시그모이드 활성함수를 통해서 아웃풋

### 활성함수 layer 층은 relu 함수를 통과
### output은 sigmoid 함수를 통과(이진 분류는 무조건 시그모이드 / 다중 분류는 soft max)

### => 히든레이어를 늘리고 싶으면 hidden_layer_sizes=(200, 10, 4) 이렇게 늘리기
