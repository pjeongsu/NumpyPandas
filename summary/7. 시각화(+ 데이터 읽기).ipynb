# 시리즈

### 시각화

```python
data = {'aa':10,'bb':20,'cc':30,'dd':40,'ee':50}
sr = pd.Series(data, name='국어점수')
sr
```

```python
# 선 그래프

sr.plot()
plt.show()
```

<pre>
kind : 그래프 종류 
figsize : 창의 크기 
title : 그래프 이름
legand : 범례 유무 
grid : 보조선 유무
ylim : y축의 범위 
rot : x축 글자가 기울어짐 정도
</pre>

```python
# 바 차트

sr.plot(kind='bar', 
        figsize=(10, 8), 
        title='성적 데이터', 
        legend=True, 
        grid=True, 
        ylim=(0, 100), 
        rot=0)

plt.show()
```

```python
# 수평바 차트

sr.plot(kind='barh',
        figsize=(10, 8), 
        title='성적 데이터', 
        legend=True, 
        grid=True, 
        rot=0)

plt.show()
```

```python
# 히스토그램 / bins 는 계급 수 설정

sr.plot(kind='hist', bins=5)
plt.show()

# pd.cut(sr, 5).value_counts().sort_index()
```

```python
# 히스토그램 / bins를 임의로 줘도 됨

sr.plot(kind='hist', bins=[0, 20, 40, 60]) 
plt.show()
```

```python
# .hist라고 적어도 됨

sr.hist(bins=[0, 20, 40, 60])
plt.show()
```

```python
# 파이 차트

sr.plot(kind='pie', autopct='%.2f') # autopct : 비율지정 - 소수 둘째자리까지
plt.show()
```

```python
# 박스plot

sr.plot(kind='box')
plt.show()
```

# 데이터프레임

```python
data =  {'eng':[10,30,50,70],
         'kor':[20,40,60,80],
         'math':[90,50,20,70]}
df = pd.DataFrame(data,index=['a','b','c','d'])
df
```

```python
# 선 그래프

df.plot()
plt.show()
```

```python
# style지정 - 색, 선 종류

df.plot(kind='line', style=['ro--', 'b<-', 'gs--'])
plt.show()
```

```python
# 바 차트
 
df.plot(kind='bar')
plt.show()
```

```python
# y 속성 : 내가 원하는 칼럼만 추출 / 디폴트는 모든 칼럼

df.plot(kind='bar', y=['eng', 'kor']) 
plt.show()
```

```python
# 스택 구조의 바 차트
# 총점이 가장 높은 것을 시각적으로 알 수 있음
# plt.text로 스택 안에 글자 적기 - 칼럼 인덱스 / 수치 / 텍스트 / 센터지정

df.plot(kind='bar', stacked=True, figsize=(10,8)) 

plt.text(0, 50,  'test1', va='center', ha='center') # 인덱스가 0인 칼럼의 값 50 부분에 텍스트 넣기 
plt.text(1, 50,  'test2', va='center', ha='center') # 인덱스가 1인 칼럼의 값 50 부분에 텍스트 넣기 

plt.show()
```

```python
df.plot(kind='bar', figsize=(10,8), stacked=True)

plt.text(0, 5, '10', va='center', ha='center') 
plt.text(0, 20, '20', va='center', ha='center') 
plt.text(0, 75, '90', va='center', ha='center') 

plt.show()
```

```python
# for문을 사용해서 각각의 스택 안에 값 찍기

df.plot(kind='bar', figsize=(10, 8), stacked=True)

n = 0
for i, sr in df.iterrows():
    e = sr['eng']  
    k = sr['kor']
    m = sr['math']
    plt.text(n , e/2, e, va='center', ha='center')
    plt.text(n , e + k/2, e, va='center', ha='center')
    plt.text(n , e + k + m/2, e, va='center', ha='center')
    n+=1
    
plt.show()
```

```python
df.plot(kind='bar', stacked=True)

n = 0
for i, sr in df.iterrows():
    e,k,m = (sr['eng'], sr['kor'], sr['math'] )
    plt.text(n,e/2,f'{e}점', va='center', ha='center')
    plt.text(n,e+k/2,f'{k}점', va='center', ha='center')
    plt.text(n,e+k+m/2,f'{m}점', va='center', ha='center')
    n+=1

plt.show()
```

```python
# 스카터 차트 : 칼럼 두 개를 비교

# x : 칼럼명
# y : 칼럼명
# size : 원 사이즈
# c : 원 색깔
# 색상 : r(빨간색), g(녹색), b(파란색), k(검은색), m(자주색), y(노란색), c(민트색)

df.plot(kind='scatter', 
        x='eng',
        y='math',
        s=[100,200,300,400],
        c=['r','b','g','c'] )

plt.show()
```
