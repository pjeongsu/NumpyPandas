# 7. 1800년대,1900년대,2000년대 각 남아,여야수 평균을 구하시요

```python
boys_1800_mean = birthDF[(birthDF.index>=1800) & (birthDF.index<1900)]['남아수'].mean()
girls_1800_mean = birthDF[(birthDF.index>=1800) & (birthDF.index<1900)]['여아수'].mean()
boys_1900_mean = birthDF[(birthDF.index>=1900) & (birthDF.index<2000)]['남아수'].mean()
girls_1900_mean = birthDF[(birthDF.index>=1900) & (birthDF.index<2000)]['여아수'].mean()
boys_2000_mean = birthDF[(birthDF.index>=2000)]['남아수'].mean()
girls_2000_mean = birthDF[(birthDF.index>=2000)]['여아수'].mean()

print(f'1800년대 남아수 평균 : {boys_1800_mean}, 1800년대 여아수 평균 : {girls_1800_mean}')
print(f'1900년대 남아수 평균 : {boys_1900_mean}, 1900년대 여아수 평균 : {girls_1900_mean}')
print(f'2000년대 남아수 평균 : {boys_2000_mean}, 2000년대 여아수 평균 : {girls_2000_mean}')
```

```python
y1800 = birthDF.loc[1800:1899]
y1900 = birthDF.loc[1900:1999]
y2000 = birthDF.loc[2000:]

df = pd.DataFrame([y1800.mean().round(2), y1900.mean().round(2), y2000.mean().round(2)])
df.index = ['1800년대', '1900년대', '2000년대']
df.index.name = '평균값'
df.T
```

# 11. 2000년 이후 남아수 여아수를 바차트로 나타내시오

```python
birth_21c =birthDF[birthDF.index >=2000]
birth_21c.plot( kind = 'bar', stacked=True, legend= None)  # stacked 를 주며 총점수 알수있음

n = 0
for i, sr in birth_21c.iterrows():
    b , g = ( sr['남아수'], sr['여아수'])
    plt.text(n,b/2,'남' ,va='center', ha='center',)
    plt.text(n,b+g/2,'여' ,va='center', ha='center')
    n +=1
plt.show()
```
