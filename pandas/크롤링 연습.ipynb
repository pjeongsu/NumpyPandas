```python
import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
import urllib.request as REQ

import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.size"] = 20
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
curl = 'http://ncov.mohw.go.kr/bdBoardList_Real.do?brdId=1&brdGubun=13&ncvContSeq=&contSeq=&board_id=&gubun='
response = REQ.urlopen(curl)
soup = BeautifulSoup(response, 'html5lib') 
soup
```

```python
ulTag = soup.select('table',{'class':'num midsize'})
ulTag
```

```python
name=[]

for l in ulTag:
    for trs in l.select('tr'):
        print(trs.select_one('th').string)
```

```python
# 데이터 중 nan 값이 있으면 select_one으로 하면 오류가 뜸 -> 따라서 select로 빼낸 뒤 다시 for문을 사용

for l in ulTag:
    for trs in l.select('tr'):
            for a in trs.select('td:nth-of-type(4)'):
                print(a.string)
```
