```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

plt.rcParams["font.size"] = 20
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
# matplotlib를 직접 이용
# plot은 라인차트 전용

# x축을 지정을 안해주면 0부터~ 범위

y = [10,14,19,20,25]
plt.plot(y)
plt.show() # 비디오 메모리에 전송
```

```python
# x를 기준으로 x가 1일 때 y1이 10, y2가 5 ///

# 선 스타일 줄 수 있음 - 칼러 / 마커 / 라인종류

x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.plot(x, y1, 'bo-', x, y2, 'rs--',x, y3, 'yp-.')
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]

# plt.figure(figsize=(6, 4)) # 디폴트 창 - 기본으로 존재함(사이즈는 6, 4)
plt.plot( x,y1,'bo-' ) 
plt.figure(figsize=(4, 4)) # 창을 그려줌
plt.plot( x,y2,'rs--')
plt.plot( x,y3,'yp-.')
plt.show()
```

```python
# 창의 디폴트사이즈는 6, 4
# subplot 창을 나눌 때 사용

x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.subplot(2, 2, 1) # 행, 열, 넘버
plt.plot(x, y1)
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.show()
```

```python
# subplots

x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
fig, axes = plt.subplots(2, 2, figsize=(7, 7)) # 창에 대한 객체를 반환 
axes[0, 0].plot(x, y1) # axes 넘파이 배열로 되어 있음
axes[0, 1].plot(x, y2)
axes[1, 0].plot(x, y3)
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
plt.xlabel('x축라벨')
plt.ylabel('y축라벨')
plt.grid(True)
plt.title('타이틀')
plt.plot(x, y1, 'bo-', x, y2, 'rs--',x, y3, 'yp-.')
plt.legend(['국어', '영어', '수학']) # 범례는 plot 이전에 적어주면 안됨
plt.show()
```

```python
xname = ['홍길동','이순신','임꺽정','김철수','이황']
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
plt.xlabel('이름')
plt.ylabel('점수')
plt.title('성적')

plt.grid(True)
plt.xticks(x, xname, rotation=45) # x를 xname으로 표시해라! -> 1 대신에 홍길동으로 표시됨

plt.plot(x,y1,'ro--', alpha=0.5) # alpha는 차트를 반투명하게
plt.text(1, 10, '10점')
plt.text(2, 14, '14점')

for n1, n2 in zip(x, y1):
#     print(n1, n2)
    plt.text(n1, n2, f'{n2}점')

plt.legend(['국어'])
plt.show()
```

```python
x = [1,2,3,4,5]
y1 = [10,14,19,20,25]
y2 = [5,8,10,25,20]
y3 = [30,5,20,11,15]
fig, ax1 = plt.subplots() # 디폴트는 하나의 창 1, 1, 1
ax1.set_xlabel('x축') # axis의 함수는 xlabel이 아니라 set_xlabel
ax1.set_ylabel('y1축', color='blue')
ax1.plot(x, y1)

ax2 = ax1.twinx() # 똑같은 ax1창을 반환하는데 축을 오른쪽으로 잡음
ax2.plot(x, y2, 'r--')
ax2.set_ylabel('y2축', color='red', rotation=0)

plt.show()
```
