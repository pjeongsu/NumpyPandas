# 1. 3개의 데이터를 병합하시요.

```python
df = pd.read_csv('../data/2014년 졸음운전 교통사고.csv',encoding='euc-kr')
df1 = pd.read_csv('../data/2015년 졸음운전 교통사고.csv',encoding='euc-kr')
df2 = pd.read_csv('../data/2016년 졸음운전 교통사고.csv',encoding='euc-kr')
```

```python
df4 = pd.concat( [df,df1,df2])
df4
```

```python
df4 = pd.concat( [df,df1,df2]).reset_index(drop=True)
df4
```

```python
df4.set_index('구분', inplace=True)
df4
```

# 2. 년도, 월별로 인덱스를 설정(멀티인덱스) 하고 데이터를 보여 주시요

```python
# 멀티 인덱스는 set_index

df5= df4.copy()
df5.index = pd.to_datetime(df5.index, format='%Y년%m월')
```

```python
df5.set_index([df5.index.year, df5.index.month])
```

# 3. 년도 및 월별 평균 사망자를 보여주시요

```python
df5.groupby([df5.index.year, df5.index.month]).mean()[['사망(명)']]
```

# 4.  2016년 사고대비 사망율을 구하시요

```python
# 인덱스 문자열

ratio_2016 = df4[df4.index.str.contains('2016년')]
ratio_2016['사고대비 사망율'] = ((ratio_2016['사망(명)'] / ratio_2016['사고(건)'])*100).round(2)
ratio_2016
```

```python
# 인덱스 시계열

df5['사고대비 사망율'] = ((df5['2016']['사망(명)'] / df5['2016']['사고(건)'])*100).round(2)
df5['2016']
```

# 5. 2014년도 월별 사망, 부상 데이터를 바차트로 보여주시요

```python
# 인덱스 문자열

df4[df4.index.str.contains('2014년')].plot(kind='bar', y=['사망(명)', '부상(명)'], figsize=(8, 6))
```

```python
# 인덱스 시계열

df6 = df5['2014'][['사망(명)','부상(명)']]
df6.head()
df6.plot(kind='bar', rot=45)
```

# 6. 2015년 대비  사망이 가장 많이 증가한 2016년도 월을 구하시요

```python
sr = df5['2016']['사망(명)']-df5['2015']['사망(명)'].values
sr
```

```python
sr.nlargest(1, keep='all')
```
