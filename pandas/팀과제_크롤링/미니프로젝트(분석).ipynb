```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
from datetime import datetime 
from pandas_datareader import data

matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus'] = False

plt.rcParams["font.size"] = 12
```

```python
df = pd.read_csv('merge.csv',index_col=False,thousands=',')
df
```

```python
# 필요없는 열 삭제
df.drop('Unnamed: 0',axis=1, inplace=True)
```

```python
df= df.sort_values('지역')
```

```python
df = df.set_index(['지역','국가'])
```

### 아메리카

```python
ame = df.loc['아메리카']
ame
# for a,b,c in ame['']
```

```python
Cp = []
for n1, n2 in zip(ame['인구수'],ame['환자발생수']):
    Cp.append(round(n2/n1,2)*100)
Cp
```

```python
plt.figure(figsize=(20,12))
plt.xticks(rotation=90)
plt.title('')
plt.xlabel('국가')
plt.ylabel('인구대비 확진율')
plt.plot(ame.index,Cp)
```
