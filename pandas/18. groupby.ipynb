```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rcParams['axes.unicode_minus']=False
plt.rcParams["font.size"] = 20
plt.rcParams["figure.figsize"] = (8, 6)
```

```python
df =pd.read_excel('../data/hotel.xlsx', index_col='Unnamed: 0')
df
```

```python
df['grade'].unique()
```

```python
df['grade'].value_counts().sort_index()
```

```python
df.info()
```

```python
# 등급별 가격의 합(~별)
# DataFrameGroupBy (집계관련함수 : sum, mean, count, max, min, agg)

g = df.groupby('grade')
g
```

```python
g.sum()[['price']] # 집계가능(숫자) 모든 컬럼
```

```python
# 혹은
df.groupby('grade', as_index=False)[['price']].sum()
```

```python
g.mean()[['price']].round(2)
```

```python
g.max()[['price']]
```

```python
g.min()[['price']]
```

```python
# 개수

g.count()[['price']]
```

```python
# 여러 개의 집계 함수나 사용자 정의 함수를 쓰고 싶다면 -> agg함수 사용

mydf = g.agg(['sum', 'mean'])['price']
mydf.coulumns=['가격의 총합', '가격의 평균']
mydf
```

```python
# 혹은

df.groupby('grade')['price'].agg(['sum', 'mean'])
```

```python
df.groupby('grade')[['price']].agg(['sum', 'mean'])
```
